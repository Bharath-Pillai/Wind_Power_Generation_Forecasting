```python
# Installing necessary libraries
!pip install pandas numpy scikit-learn matplotlib seaborn

# Importing required libraries
import pandas as pd  # Data manipulation
import numpy as np  # Numerical computations
import matplotlib.pyplot as plt  # Visualization
import seaborn as sns  # Enhanced data visualization
from sklearn.model_selection import train_test_split  # Splitting data
from sklearn.preprocessing import StandardScaler  # Data scaling

# Load the CSV files (Dataset)
l1 = pd.read_csv('Location1.csv')
l2 = pd.read_csv('Location2.csv')
l3 = pd.read_csv('Location3.csv')
l4 = pd.read_csv('Location4.csv')

# Display first few rows of Location 1 dataset
print(l1.head())

# Add a new column to specify the location
l1['Location'] = 'Location1'
l2['Location'] = 'Location2'
l3['Location'] = 'Location3'
l4['Location'] = 'Location4'

# Concatenating all location datasets into a single DataFrame
data_combined = pd.concat([l1, l2, l3, l4], ignore_index=False)

# Display first few rows of merged dataset
print(data_combined.head())

# Save the merged dataset to a new CSV file
data_combined.to_csv('merged_locations.csv', index=True)

# Display basic information about the dataset
data_combined.info()

# Display dataset statistics
print(data_combined.describe().T)

# Check for missing values
print(data_combined.isnull().sum())

# Check for duplicate rows
print("Number of duplicate rows:", data_combined.duplicated().sum())

# Encode categorical variable 'Location' using one-hot encoding
data_combined = pd.get_dummies(data_combined, columns=['Location'], drop_first=True)

# Remove the 'Time' column
data_combined.drop(columns=['Time'], inplace=True)

# Display the updated dataset structure
print(data_combined.head())
```
